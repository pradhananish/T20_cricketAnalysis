```python
import pandas as pd
import json

with open('t20_wc_match_results.json') as file:
    data = json.load(file)
data
df_match=pd.DataFrame(data[0]['matchSummary'])
df_match.head()
#df.shape
```

```python
df_match.rename({'scorecard':'match_id'}, axis=1, inplace=True)
df_match.head()
df_match[['team1','team2']].astype(str)
df_match.head()
df_match.info()
```

```python
match_ids_dict = {}
for _, row in df_match.iterrows():
    team1 = row['team1'].strip().lower()
    team2 = row['team2'].strip().lower()
    key1 = f"{team1} vs {team2}"
    key2 = f"{team2} vs {team1}"
    match_ids_dict[key1] = row["match_id"]
    match_ids_dict[key2] = row["match_id"]
match_ids_dict
```

```python
query_key = "Namibia vs Sri Lanka".lower()
match_ids_dict[query_key]
```

batting summary 

```python
with open('t20_wc_batting_summary.json') as file:
    data=json.load(file)

    all_records=[]
    for rec in data:
        all_records.extend(rec['battingSummary'])
df_batting= pd.DataFrame(all_records)
df_batting.head()
df_batting.tail()
```

```python
df_batting["out/not_out"]=df_batting.dismissal.apply(lambda x:"out" if len(x)>0 else "not_out")
df_batting.head(15)
```

```python
df_batting.drop(columns=["dismissal"], inplace=True)
df_batting.head(11)
```

```python
df_batting['batsmanName']=df_batting['batsmanName'].apply(lambda x: x.replace('\aθ',''))
df_batting.head()
```

```python
df_batting['batsmanName']=df_batting['batsmanName'].astype(str).str.replace("€", "")
df_batting.head(19)
```

```python
df_batting['batsmanName']=df_batting['batsmanName'].astype(str).str.replace("†", "")
df_batting.tail(19)
```

```python
query_key = "Namibia vs Sri Lanka".lower()
match_ids_dict[query_key]
```

```python
df_batting["match"] = df_batting["match"].str.strip().str.lower()
df_batting["match_id"] = df_batting["match"].map(match_ids_dict)
df_batting
```

```python
import os

# Create the folder if it doesn't exist
os.makedirs("t20 cricket", exist_ok=True)
df_batting.to_csv("t20 cricket/t20_cricket_clean_data.csv", index=False)
```
